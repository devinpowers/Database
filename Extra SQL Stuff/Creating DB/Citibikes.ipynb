```python
db = sqlite3.connect('citi.db')
```

```python
print(pandas.read_sql_query('select count(*) from Citibikes', db))
```

```python
res = pandas.read_sql_query('select * from CitiBikes', db, chunksize =100_000)
```

```python
res
```

```python
def Q(sql):
    res = pandas.read_sql_query(sql, db, chunksize = 100_000)
    return next(res)
```

```python
Q("select * from Citibikes")
```
